```python
# station_files 변수에 각 엑셀 파일의 경로와 이름을 리스트로 저장 
station_files = glob("C:/Users/Sung/Desktop/Opinet_Data/세차장/지역*.xls")
len(station_files)
```

```python
station_files[0]
```

```python
# 파일 한개 가져오기 
tmp_raw = []
tmp = pd.read_excel('C:/Users/Sung/Desktop/Opinet_Data/세차장\\지역_위치별(주유소) (1).xls', header = 2) # header 는 3번째 줄에 있음 
tmp_raw.append(tmp)
tmp_raw
```

```python
# 파일 여러개 가져오기
# read_excel 로 각 파일을 읽은 후
# tmp_raw 변수에 append
# 반복문 끝나면 pd.concat 명령으로 쉽게 하나로 합침
# 25개 엑셀 파일 하나로 합침

tmp_raw = []
for file_name in station_files:
    tmp = pd.read_excel(file_name, header = 2) # header 는 3번째 줄에 있음 
    tmp_raw.append(tmp)

station_raw = pd.concat(tmp_raw)
```

```python
station_raw.info() # 가격 정보가 숫자형(int, float) 아니어서 나중에 처리해야 함
```

```python
# 원하는 컬럼 가지고 오고 이름도 다시 정의 stations 라는 변수에 저장 
# 지역, 상호, 주소, 상표, 셀프여부 
stations = pd.DataFrame({"sido" : station_raw["지역"],
                         "os_nm": station_raw["상호"],
                         "new_adr": station_raw["주소"],
                         "oil_price": station_raw["휘발유"],
                         "poll_div_cd": station_raw["상표"],
                         "self_yn": station_raw["셀프여부"]})
stations.head()
```

## 가격 정보 전처리

```python
stations[stations["oil_price"] == "-"]
```

```python
# * 2. 정보가 기입되지 않은 주유소는 대상에서 제외 
stations_notnull = stations[stations["oil_price"] != '-']
stations_notnull.head()
stations_notnull["oil_price"] = [float(value) for value in stations_notnull["oil_price"]]
stations_notnull.info() # 가격 정보가 float 으로 바뀜\
stations_notnull.reset_index(inplace = True) # 인덱스를 처음부터 다시 기록하도록 함 
del stations_notnull["index"] # index 라는 컬럼이 하나 더 생성되어, 그 부분 제거

# 구별로 가격 평균 구하기 
import numpy as np 
si_gu_data =pd.pivot_table(stations_notnull, index =["si","gu"] , values = ["oil_price"],aggfunc = np.mean)
```

```python
si_gu_data.reset_index(inplace = True) # 인덱스를 처음부터 다시 기록하도록 함 
```
