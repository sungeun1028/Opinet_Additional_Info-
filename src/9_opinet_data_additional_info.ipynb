```python
import pandas as pd 
carwash = pd.read_csv("C:/Users/Sung/Desktop/Opinet_Data/주유소_부가정보/carwash.csv", encoding= "cp949")
repair = pd.read_csv("C:/Users/Sung/Desktop/Opinet_Data/주유소_부가정보/repair.csv", encoding= "cp949")
conv_store = pd.read_csv("C:/Users/Sung/Desktop/Opinet_Data/주유소_부가정보/conv_store.csv", encoding= "cp949")
all247 = pd.read_csv("C:/Users/Sung/Desktop/Opinet_Data/주유소_부가정보/all247.csv", encoding= "cp949")
```

```python
merge_dat1 = pd.merge(carwash, repair, how = "outer",
                     left_on = ["sido", "si", "gu","os_nm","new_adr","api_adr","poll_div_cd","self_yn"], 
                     right_on = ["sido", "si", "gu","os_nm","new_adr","api_adr","poll_div_cd","self_yn"])
```

```python
merge_dat1
```

```python
merge_dat2 = pd.merge(conv_store, all247, how = "outer",
                     left_on = ["sido", "si", "gu","os_nm","new_adr","api_adr","poll_div_cd","self_yn"], 
                     right_on = ["sido", "si", "gu","os_nm","new_adr","api_adr","poll_div_cd","self_yn"])
```

```python
merge_dat2
```

```python
merge_dat3 = pd.merge(merge_dat1, merge_dat2, how = "outer",
                     left_on = ["sido", "si", "gu","os_nm","new_adr","api_adr","poll_div_cd","self_yn"], 
                     right_on = ["sido", "si", "gu","os_nm","new_adr","api_adr","poll_div_cd","self_yn"])
```

```python
merge_dat3
```

```python
merge_dat3 = merge_dat3.fillna("N")
```

```python
oil_add_info = merge_dat3.copy()
```
